```python
import json
import pandas as pd
```

```python
with open('skip_2.json') as f: 
    data_old=json.load(f)
with open('skip_3.json') as f: 
    data_new=json.load(f)
# print(data)
```

```python
df_old=pd.DataFrame(data_old['bikes'])
df_new=pd.DataFrame(data_new['bikes'])
```
